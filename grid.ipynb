```python
from pandas_datareader import data as pdr
from datetime import date
import yfinance as yf
yf.pdr_override()
import pandas as pd
import numpy as np



today = date.today()
ticker1 = 'SPY'
start_date = '2020-01-02'
end = today
df = pdr.get_data_yahoo(ticker1, start=start_date, end=today)
df
```

```python
#[id,date,priceOpen,priceClose]
orders = []
a=0
strDate = str(df.index[a]).split()[0]
orders.append([a,strDate, df.Close[a], 0])
lastPrice = df.Close[a]
grid=5

a += 1
while a<len(df.iloc[1:]):
    spot = df.Close[a]
    if abs(spot-lastPrice) > grid:
        strDate = str(df.index[a]).split()[0]
        orders.append([a,strDate, spot, 0])
        lastPrice = spot
        for idx, item in enumerate(orders):
            if item[3]==0 and item[2]+grid <= spot:
                orders[idx] = [item[0],item[1],item[2], spot]
    a += 1
    
```

```python
orders
```

```python
profits = []
for order in orders:
    profit = order[3]-order[2] 
    profits.append(profit) if profit > 0 else None

profits, sum(profits)
```
